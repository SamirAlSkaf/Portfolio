Function for calculating y = m * x + b

```python
def get_y(m,b,x):
  y=m*x+b
  return y
```

```python
get_y(5, 10, 3) == 25
```

Function for calculating error

```python
def calculate_error(m,b,point):
  x_point, y_point = point
  y_imaginary = m * x_point + b
  distance = abs(y_imaginary - y_point)
  return distance
```

```python
print(calculate_error(1, 0, (3, 4)))
```

Data of Points (x,y)

```python
datapoints = [(1, 2), (2, 0), (3, 4), (4, 4), (5, 3)]
```

Function for calculating total errors of all points

```python
def calculate_all_error(m, b, points):
  total_error = 0
  for point in datapoints:
    point_error = calculate_error(m,b,point)
    total_error += point_error
  return total_error
```

```python
print(calculate_all_error(1, 1, datapoints))
```

Creating possible m- and b-values (Range: -10 to 10 and -20 to 20; Steps-Size: 0,1)

```python
possible_ms = [m/10 for m in range(-100,100)]
possible_bs = [b/10 for b in range(-100,100)]
```

```python
smallest_error = float("inf")
best_m = 0
best_b = 0

for m in possible_ms:
  for b in possible_bs:
    error = calculate_all_error(m,b,datapoints)
    if error < smallest_error:
      best_m = m
      best_b = b
      smallest_error = error

print(best_m, best_b, smallest_error)
```
