```python
import numpy as np
import pandas as pd

# Daten importieren
data = pd.read_csv('house_prices.csv')

# Features und Zielvariable vorbereiten
X = data.drop(['SalePrice'], axis=1).values  # Features
y = data['SalePrice'].values                 # Zielvariable

# F체ge eine Spalte f체r den Bias-Term (Intercept) hinzu
X = np.c_[np.ones(X.shape[0]), X]

# Berechnung der Transponierten von X
X_transpose = X.T

# Berechnung der Gewichtsmatrix (W)
X_transpose_X = X_transpose @ X
X_transpose_y = X_transpose @ y
W = np.linalg.inv(X_transpose_X) @ X_transpose_y

# Vorhersagen
y_pred = X @ W

# Fehlerberechnung (Mean Squared Error)
mse = np.mean((y - y_pred) ** 2)

# Zus채tzliche Ausgaben zur Modellbewertung
print('Erste 5 Vorhersagen (y_pred):', y_pred[:5])
print('Erste 5 tats채chliche Werte (y):', y[:5])
print('Mean Squared Error (MSE):', mse)
```
